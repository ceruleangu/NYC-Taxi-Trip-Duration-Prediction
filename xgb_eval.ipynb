```python
import xgboost as xgb

model_path = './numerical/0099.model'
test_data_path = './numerical/test.txt#dtrain1.cache'
```

```python
dtest = xgb.DMatrix(test_data_path)
```

```python
bst = xgb.Booster(model_file=model_path)
```

```python
bst.eval(dtest)
```

```python
y1=bst.get_score(importance_type="gain")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bst.feature_names=['passenger count','pickup longitude','pickup latitude','dropoff longitude','dropoff latitude','snow','rain', 'haversine distance','manhantan distance', 'bearing','vendor id','payment type','pickup location id','dropoff location id','is holiday', 'week of the year', 'quarter of the day']
bst.feature_types=['q','q','q','q','q','q','q','q','q','q','q','q','q','q','q','q','q']
```

```python
xgb.plot_importance(bst,importance_type='weight',xlabel="Feature Importance Score",title="")
```
